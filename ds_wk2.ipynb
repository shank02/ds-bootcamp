Define two custom numpy arrays, say A and B. Generate two new numpy
arrays by stacking A and B vertically and horizontally.

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])


vertical_stack = np.vstack((A, B))
horizontal_stack = np.hstack((A, B))

print("Vertical Stack: \n", vertical_stack)
print("Horizontal Stack:\n", horizontal_stack)
```

Find common elements between A and B

```python
common_elements = np.intersect1d(A, B)
print("common elements between A and B: ", common_elements)
```

Extract all numbers from A which are within a specific range. eg
between 5 and 10

```python
A = np.array([1, 3, 5, 7, 9, 11, 13])
#extract no.s between 5 and 10
result = A[(A >= 5) & (A <= 10)]
print("Numbers between 5 and 10:", result)
```

Filter the rows of iris_2d that has petallength (3rd column) > 1.5
and sepallength (1st column) < 5.0

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
filtered_iris = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0)]
print("Filtered rows - \n", filtered_iris)
```

How to get the rows of a dataframe with row sum > 100?

```python
#df with random integers
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
```

```python
#rows with sum >100
filtered_rows = df[df.sum(axis=1) > 100]
print(filtered_rows)
```
